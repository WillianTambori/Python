###1. Construir um algoritmo para calcular e apresentar o total de salários pagos de uma, mas não é informado a quantidade de pessoas, então use como critério de parada (condição da estrutura de repetição, digitar zero no salário para sair.


```
salario = float(input("informe o salário :"))
total = 0
while salario > 0:
    total = total + salario
    salario = float(input("informe o salário...."))
print("o total de salário pagos é :",total)
```

###2. Construir um algoritmo para calcular e apresentar a idade atual de algumas pessoas em relação ao ano atual, mas não é informado a quantidade de pessoas, então use como critério de parada (condição da estrutura de repetição, digitar zero no ano de nascimento para sair.

```
ano= int(input("informe o seu ano de nascimento:"))
from datetime import datetime
hoje = datetime.now()
ano_atual = hoje.year
while ano >0:   
    idade = ano_atual - ano
    print("sua idade atual é:",idade)
    ano = int(input("informe o ano de nascimento :"))
print('obrigado por colaborar')
```

###3. Faça um programa que receba a idade e a altura de várias pessoas. Calcule e exiba a média das alturas das pessoas com mais de 20 anos. Para encerrar a entrada de dados, digite uma idade negativa ou igual a zero.




```
idade = int(input("informe a idade :"))
soma_altura =0
contador = 0
while idade > 0:
    if idade >20:
        alturas = float(input("informe a altura:"))
        soma_altura = soma_alturas + alturas
        contador = contador + 1
    idade = int(input(" informe a idade.."))

media = soma_alturas / contador
print("A média de altura com mais de 20 anos é", media)
```

###5. No final do ano muitas pessoas compram presentes. Faça um programa que registre alguns dados das pessoas, usando como critério de parada a letra ‘n’, para a pergunta “Deseja cadastrar outro (‘s’/’n’)?”, para identificar o perfil dos compradores numa loja de roupas e apresente como resultado a:
*   Quantidade de mulheres e de homens;
*   Quantidade de mulheres e de homens abaixo e acima de 18 anos.

```
cadastro = input("deseja entrar no cadastrar? digite 'S' para sim 'N' para não : ")
masculino = 0
feminino = 0
maior = 0 
menor = 0
while cadastro != "N" and cadastro != "n":
    nome = input("informe o nome: ")
    sexo = input("infome o sexo: ")
    idade = int(input("informe a idade: "))
    if sexo == "M" or sexo == "m":
        masculino = masculino + 1
    if sexo == "F" or sexo =="f":
        feminino += 1
    if idade >= 18:
        maior += 1
    if idade < 18:
        menor += 1    
    cadastro = input(" S sim N para não :")
print(f" quantidadades de maiores de idade: {maior}")
print(f" quantidadades de menores de idade: {menor}")
print(f" quantidadades de masculinos : {masculino}")
print(f" quantidadades de feminino: {feminino}")
```
