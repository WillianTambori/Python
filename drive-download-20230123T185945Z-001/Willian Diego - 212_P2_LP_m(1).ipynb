###**1.** [6,0 pontos] Em Python crie um programa que atenda o menu a seguir. Desenvolva uma função para cada opção do menu. Utilize a criação de estruturas. Cadastrar 5 (vetor) vendas ( Tipo_Venda: cod_venda (int), dia(int), total(float), cliente (Tipo_Cliente: cod_cliente (int), nome(str) )

Menu de opções:

1 Cadastrar clientes

2 Mostrar todos os clientes

3 Cadastrar vendas

4 Mostrar todas as vendas

5 Mostrar o total de vendas de um determinado dia

6 Armazenar todos os campos da venda em um arquivo

7 Apresentar o conteúdo do arquivo

8 Sair


**Observaçôes**: 

* Na opção 5, o usuário do sistema informará/digitará o dia que quer e o sistema/programa fará uma pesquisa e calculará o total de vendas de dias iguais, somar o total dos dois lançamentos do dia 25, três no dia 10, lembrando que, as vendas podem ter o registro como demonstra este exemplo, uma venda dia 10, outra dia 24, duas no dia 25, uma dia 20, uma dia 10, mais uma outra no dia 10 e assim por diante.

* Na opção 6 e 7 o nome do arquivo deverá ser **vendas.txt**

```
class client:
    kind = 0
    code = 0
    name = ""
class sale:
    kind = 0
    code = 0
    day =  0
    value = 0.0
    account = client()
def register_client(p):
    option = int (input("informe o tipo de cliente ou '0' para sair do cadastro"))
    while option > 0:
        vet = client()
        vet.kind = option
        vet.code = int (input('insira o código do cliente: '))
        vet.name = input('insira o nome do cliente: ')
        option = int (input("informe o tipo de cliente ou '0' para sair do cadastro: "))
        p.append(vet)
    return p
def pick_account(b):
    c = 1
    while c > 0:
        a = int (input("informe o código do cliente :"))
        for i in range(len(b)):
            if a == b[i].code:
                return b[i].name
                c= c - 1
                break
        print("codigo de cliente não encontrado")         

def register_sale(p,s):
    option = int (input("informe o tipo de venda ou '0' para sair do cadastro: "))
    while option > 0:
        vet = sale()
        vet.kind = option
        vet.code = int (input('insira o código da venda: '))
        vet.day = input('insira o dia da venda: ')
        vet.value = float(input('insira o valor da venda: '))
        vet.account = pick_account(s)
        option = int (input("informe o tipo de venda ou '0' para sair do cadastro: "))
        p.append(vet)        
    return p

def show(x):
    o = ["tipo de cliente     ","código do cliente   ","nome do cliente     "]
    print("{0:<20}".format(o[0]),end="")
    print("{0:<20}".format(o[1]),end="")
    print("{0:<40}".format(o[2]))
    for i in range(len(x)):
        print("{0:<20}".format(x[i].kind),end="")
        print("{0:<20}".format(x[i].code),end="")
        print("{0:<40}".format(x[i].name))

def show_sale(x):
    o = ["tipo de venda","código da venda","dia da venda","valor da venda","cliente"]
    print("{0:<20}".format(o[0]),end="")
    print("{0:<20}".format(o[1]),end="")
    print("{0:<20}".format(o[2]),end="")  
    print("{0:<20}".format(o[3]),end="")
    print("{0:<40}".format(o[4]))
    for i in range(len(x)):
        print("{0:<20}".format(x[i].kind),end="")
        print("{0:<20}".format(x[i].code),end="")
        print("{0:<20}".format(x[i].day),end="")
        print("{0:<20}".format(x[i].value),end="")
        print("{0:<40}".format(x[i].account))

def show_sale_day(y):
    c = 0
    date = int(input("informe o dia da pesquisa: ")) 
    for i in range(len(y)):
        if y[i].day == date:
            c = c + y[i].value
    print(" o total de vendas do dia ",date," é : ",c)

def files(c):
    p2 = open('vendas.txt','w')
    for i in range(len(c)):
        t = sale()
        t.kind = c[i].kind
        t.code = c[i].code
        t.day = c[i].day
        t.value = c[i].value
        t.account = c[i].account
        p2.write(f'{t.kind},{t.code},{t.day},{t.value:.2f},{t.account}\n')
    p2.close()

def show_files(a):
    x = open("vendas.txt","r")
    o = ["tipo de venda","código da venda","dia da venda","valor da venda","cliente"]
    print("{0:<20}".format(o[0]),end="")
    print("{0:<20}".format(o[1]),end="")
    print("{0:<20}".format(o[2]),end="")  
    print("{0:<20}".format(o[3]),end="")
    print("{0:<40}".format(o[4]))
    for i in x.readlines():
        a,b,c,d,e = i.strip().split(',')
        print("{0:<20}".format(a),end="")
        print("{0:>20}".format(b),end="")
        print("{0:>20}".format(c),end="")
        print("{0:>20}".format(d),end="")
        print("{0:>40}".format(e))
    x.close()

def menu():
    A = ["Cadastrar clientes","Mostrar todos os clientes","Cadastrar vendas","Mostrar todas as vendas","Mostrar o total de vendas de um determinado dia","Armazenar todos os campos da venda em um arquivo","Apresentar o conteúdo do arquivo","Sair"]
    print("*** opções disponivel ***")
    for i in range(len(A)):
        print(f"{i+1}. {A[i]}")
    print("*************************")
    o = int(input("informe a opção desejada.."))
    return o
def main():
    a = []
    b = []
    op = menu()
    while op >= 1:
        if op == 1:
           register_client(a)
        elif op == 2:
           show(a) 
        elif op == 3:
           register_sale(b,a)  
        elif op == 4:
           show_sale(b) 
        elif op == 5:
           show_sale_day(b) 
        elif op == 6:
            files(b)
        elif op == 7:
            show_files(b) 
        elif op == 8:
            break
        else:
            print("informe uma das opções")
        op = menu()

    print('\nTermino do programa...............')



main()    
```
