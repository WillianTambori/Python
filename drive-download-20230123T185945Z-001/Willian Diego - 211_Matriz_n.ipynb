### 1. Faça um programa que leia uma matriz 2x3 (2 linhas, 3 colunas). Apresenta os elementos da matriz e seus respectivos índices.

```
matriz = [] # ................................................... Lendo a matriz
for lin in range(2): # lin corresponde ao índice das linhas
    vet_linha = []
    for col in range(3): # col cooresponde ao índice das colunas
        vet_linha.append(int(input('Digite o valor ['+ str(lin) + '][' + str(col) + ']: ')))
    matriz.append(vet_linha)

for lin in range(2):  # .................................. Apresentando a matriz
    for col in range(3): 
        print(matriz[lin][col], end='\t')
    print()

for lin in range(len(matriz)):  # a função len(matriz), possui a quantidade de linhas
    for col in range(len(matriz[0])): # a função len(matriz[0]), possui a quantidade de colunas
        print('O elemento ',matriz[lin][col], ' está na linha',lin,' e na coluna',col)
```

### 2. Faça um programa que carregue uma matriz 2 x 2, que representa o salário de 4 funcionários, calcule e mostre a soma total de todos os elementos que será o montante pago pela empresa a esses funcionários.

```
soma = 0
matriz=[]
for lin in range(2):
    vet_linha=[]
    for col in range(2):
        vet_linha.append(int(input("digite o valor :")))
    matriz.append(vet_linha)
for lin in range(2):
    for col in range(2):
        soma += matriz[lin][col] 
for lin in range(len(matriz)):
    for col in range(len(matriz[0])):
        print(matriz[lin][col],end="\t")
print()
print("A soma dos produtos eaquivale a",soma)
```

### 3. Faça um programa que carregue uma matriz 3 x 2, que representa preços de produtos, crie OUTRA matriz que armazene todos os preços com 7% de aumento.

```
matriz=[]
for caixa in range(3):
    vet_caixa=[]
    for prod in range(2):
        vet_caixa.append(float(input("digite o valor do produto: ")))
    matriz.append(vet_caixa)
reajuste=[]
for caixa in range(3):
    nov_caixa=[]
    for prod in range(2):
        nov_caixa.append((matriz[caixa][prod] * 1.07))
    reajuste.append(nov_caixa) 
for lin in range(3):
    for col in range(2):
        print("o produto de valor ",matriz[lin][col]," foi reajustado para",reajuste[lin][col],end="\t")
    print()
        
```

### 4. Faça um programa para armazenar em uma matriz 5x2 preços. Encontre e apresente os ÍNDICES dos valores menores que 23 reais.

```
matriz=[]
for caixa in range(5):
    vet_caixa=[]
    for prod in range(2):
        vet_caixa.append(int(input("digite o valor do produto: ")))
    matriz.append(vet_caixa)
for lin in range(5):
    for col in range(2):
        print(matriz[lin][col],end="\t")
    print()
for caixa in range(5):
    for prod in range(2):
        if matriz[caixa][prod] < 23:
            print(' produto menor de 23$ no valor de: ',matriz[caixa][prod], ' está na linha',caixa,' e na coluna',prod)
```

### 5. Faça um programa que leia números inteiros m e n e os elementos de uma matriz A de números inteiros de dimensão m x n e conte o número de elementos que são iguais a zero.


```
cont =  0
matriz=[]
m = int(input("fale o numero de linha que vc deseja: "))
n = int(input("fale o numero de coluna que vc deseja: "))
for lin in range(m):
    vt_linha = []
    for col in range(n):
        vt_linha.append(int(input("informe um numero inteiro: ")))
    matriz.append(vt_linha)
for lin in range(len(matriz)):
    for col in range(len(matriz[0])):
        if matriz[lin][col] == 0:
            cont += 1
print("numeros de elementos iguais a '0' são: ",cont)        
```

### 6. Faça um programa que carregue: *um vetor com oito posições com os nomes das lojas; * um outro vetor com quatro posições com os nomes dos produtos; * uma matriz com os preços de todos os produtos em cada loja. O programa deve mostrar todas as relações (nome do produto – nome da loja) nas quais o preço não ultrapasse R$ 120,00. Escolha qual é o tipo da função que quer.


```
lojas = []
produto=[]
preco=[]
for i in range(4):
    lojas.append(input("informe as lojas :"))
for i in range(2):
    produto.append(input("informe os produtos:"))

for lin in range (4):
    vetor_linha = []
    for col in range(2):
        vetor_linha.append(float(input(f"informe o preço do {produto[col]} da loja {lojas[lin]} : ")))
    preco.append(vetor_linha)

for lin in range(4):
    for col in range(2):
        if preco[lin][col] < 120:
            print(f"A loja {lojas[lin]}, com  o produto {produto[col]}, custa: {preco[lin][col]:.2f}")
print()        
        
```
